#### **Transformaciones de Fuerza**

estas funciones transforman la fuerza del MR experimento al MR laboratorio, en su generalidad, se puede extender, por ejemplo para pasar del MR propio al MR experimento usando las formulas de las fuerzas paralelas y perpendiculares encontradas previamente. Forma en la cual se obtuvieron las fuerzas correspondientes a cada grafico

```python
def F_ll(F_p, v_p, u):
  """ Calculo de Fuerza paralela (x)

  F_p : vector Fuerza (vista desde MR exp)
  u   : vector velocidad del Mr exp relativa al Lab
  v'  : vector velocidad de la particula (desde MR exp)"""
  D_p = 1 / (1 + np.dot(v_p,u))
  return D_p*( F_p[0] + u * np.dot(v_p, F_p) )


def F_l_(F_p, v_p, u):
  """ Calculo de Fuerza perpendicular (y)

  F_p : vector Fuerza (desde MR exp)
  u   : vector velocidad del Mr exp relativa al Lab
  """
  D_p = 1 / (1 + np.dot(v_p,u))
  gamma_u = 1 / (1 - np.dot(u,u))**.5
  return D_p * F_p[1] / gamma_u


# Transformacion de velocidades (Lab -> Exp)
def v_p(v, u):
  """ Calculo de velocidad primada

  v : vector velocidad (desde MR lab)
  u : vector velocidad del Mr exp relativa al Lab
  """
  gamma_u = 1 / (1 - np.dot(u,u))**.5
  vxp = (v[0] - u)/(1 - u*v[0])   # componente x
  vyp = v[1]/(1 - u*v[0])/gamma_u # componente y
  return np.array([vxp,vyp])
```

Grafica interactiva, con variaciones interactivas de la fuerza (MR propio), velocidad (vista desde MR lab) y rapidez u (del experimento vista desde el MR lab). Hubo que añadir una condicional excepcion, para que el usuario no pueda introducir componentes de la velocidad de la particula mayores a la de la luz.

```python
@interact(u=(0, 0.99, 0.01), vx=(0, 0.99, 0.01), vy=(0, 0.99, 0.01),
          fx=(-2, 2, 0.1), fy=(-2, 2, 0.1),
          continuous_update=False)

def interactua(u=0.6, vx=0.8,vy=0.5, fx=1, fy=2):
  if (vx**2 + vy**2)**.5 >= 1:
      print("Error: La norma de la velocidad no puede ser mayor o igual a la de la luz.")
      return

  # VARIABLES DE INTERACCION
  v = np.array([vx, vy])       # Velocidad partícula (vista desde lab)
  F_prop = np.array([fx, fy])  # Fuerza (desde MR propio)

  # Del MR propio al MR experimento
  v_prop = np.array([0, 0])  # velocidad particula (desde MR propio)
  u_prop_exp = v_p(v, u)     # velocidad relativa Mr propio (desde MR exp)

  # Fuerza en el MR experimento
  F_ll_exp = F_ll(F_prop, v_prop, u_prop_exp)[0] # igual a primer comp
  F_l_exp = F_l_(F_prop, v_prop, u_prop_exp)

  # Fuerza en el MR lab (Del MR propio al MR lab)
  F_ll_lab = F_ll(F_prop, v_prop, v)[0]
  F_l_lab = F_l_(F_prop, v_prop, v)


  # Grafica
  fig, ax = plt.subplots(1, 3, figsize=(10, 4))

  # MR propio de la partícula
  ax[0].quiver(0, 0, F_prop[0], F_prop[1], angles='xy',
               scale_units='xy', scale=1, color='r')
  ax[0].set_title('Sistema propio de la partícula')
  ax[0].set_xlabel('F_x')
  ax[0].set_ylabel('F_y')
  ax[0].set_xlim(-2.5, 2.5)
  ax[0].set_ylim(-2.5, 2.5)
  ax[0].grid()

  # MR del experimento
  ax[1].quiver(0, 0, F_ll_exp, F_l_exp, angles='xy',
               scale_units='xy', scale=1, color='y')
  ax[1].set_title('Sistema del experimento')
  ax[1].set_xlabel('F_x')
  ax[1].set_xlim(-2.5, 2.5)
  ax[1].set_ylim(-2.5, 2.5)
  ax[1].grid()

  # MR del laboratorio
  ax[2].quiver(0, 0, F_ll_lab, F_l_lab, angles='xy',
               scale_units='xy', scale=1, color='purple')
  ax[2].set_title('Sistema del laboratorio')
  ax[2].set_xlabel('F_x')
  ax[2].set_xlim(-2, 2)
  ax[2].set_ylim(-2, 2)
  ax[2].grid()
  plt.tight_layout()
  plt.show()
```
