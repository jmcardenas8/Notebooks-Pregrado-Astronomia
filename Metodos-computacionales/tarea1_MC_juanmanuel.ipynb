```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# **PUNTO 1**: Sucesion de fibonaci

```python
def fibonacci(n):
  f0=0; f1=1; i=0; l=[]
  for i in range(n):
    fn=f1+f0      #iteracion (apartir de 1+0)
    l.append(fn)  #lista elementos generados por la sucesion

    f0=f1 ; f1=fn ; i+=1  #reinicio de terminos
  return l
```

```python
n=5             #n primeros numeros de la sucesion
l= fibonacci(n) #numeros generados {1,2,3..}
l_old=[1]+l     #numeros con el primer uno: {1,1,2..}
print(l)
```

# **PUNTO 2**: grafica de la espiral de fibonacci

### Para {1,1,2,3,5...} (tomando en cuenta el primer uno)

```python
n=5
l= fibonacci(n) ; l_old=[1]+l

x0list=[0,0] ; y0list=[0,0] #centros o correccion vectores (x0, y0)
Δ=0; i=0 ; plt.figure(figsize=(8,5))

for i in range(n):
  x0=l_old[i]*np.cos(i*np.pi/2)+ x0list[i+1] #listas para x0 y y0
  y0=l_old[i]*np.sin(i*np.pi/2)+ y0list[i+1] #radio*angulo + vector anterior
  x0list.append(x0); y0list.append(y0)

  θcont= np.arange( 0+Δ , np.pi/2 + Δ , 0.01) #angulo "continuo"
  xcont= l_old[i]*np.cos(θcont) + x0list[i]    #radio*cos + correccion origen
  ycont= l_old[i]*np.sin(θcont) + y0list[i]

  plt.plot(xcont,ycont,'k') #graficacion de cada seccion
  Δ+= np.pi/2  ;  i+= 1     #cada radio barre pi/2

plt.title('Espiral de Fibonacci, n=%i' %n)
plt.grid()
```

## Para {1,2,3,5...} (misma grafica pero toma solo los elementos generados sin el primer uno)

```python
n=5
l= fibonacci(n) ; l_old=[1]+l

x0list=[0]; y0list=[0] #centros o correccion vectores (x0, y0)+primer centro
Δ=0; i=0 ; plt.figure(figsize=(8,5))

for i in range(n):
  x0= l_old[i]*np.cos((i+3)*np.pi/2) + x0list[i] #listas para x0 y y0
  y0= l_old[i]*np.sin((i+3)*np.pi/2) + y0list[i] #radio*angulo + vector anterior
  x0list.append(x0); y0list.append(y0)

  θcont= np.arange( 0+Δ , np.pi/2 + Δ , 0.01) #vectores barriendo 90°
  xcont= l[i]*np.cos(θcont) + x0list[i]       #radio*ang + correccion origen
  ycont= l[i]*np.sin(θcont) + y0list[i]

  plt.plot(xcont,ycont,'k') #grafica cada seccion
  Δ+= np.pi/2  ; i+= 1      #cada radio barre pi/2

plt.title('Espiral de Fibonacci, n=%i' %n)
#plt.plot(x0list,y0list,'o') #centros
plt.grid()
```

## **Discusion:**
<p align="justify">
Se obtuvo que la espiral para $n=5+4++4+...=5+4*m$, no cambia su comportamiento con respecto a la que se obtiene con $n=5$, es decir el ultimo radio polar graficado, termina con el mismo angulo en ambos casos, por lo que los ciclos de las espirales cumplen con mismos periodos que son multiplos de $4$. Lo que que ademas se satisface sin importar que $m$ sea par o impar. De este modo, se puede afirmar que la espiral tiene una periodicidad de $4$; luego de $4$ iteraciones la espiral vulve a tomar una forma similiar a la inicial, como si de un fractal se tratase.

Ocurre algo parecido para la suma si se toma $2*m, 6*m$ y demás pares no multiplos de $4$, con la restriccion de que $m$ tiene que ser par para que la espiral sea periodica y culmine con el mismo angulo; si no es par, esto no sucede. Ademas es claro que para los multiplos de $4$ como $8*m$ tambien se cumple la periodicidad asociada al $4$, para $m$ ya sea par o impar. </p>

# **punto 4: serie geometrica**

```python
def geometrica(r,N):
  l=[]
  #if np.abs(r)<1:
  sum=0; i=0
  for i in range(N):
    sum+= r**i
    l.append(sum)
    i+=1
  return l
  #else: print('Para r= %.1f la serie geometrica diverge' %r )
```

```python
r=0.5; N=50
x1=geometrica(r,N)
plt.plot(x1,'o-',color='darkblue')
plt.title('Serie Geometrica para $r=$%.2f y $N=$%.1f ' %(r,N))
```

```python
r=1/5; N=50
x2=geometrica(r,N)
plt.plot(x2,'o-',color='g')
plt.title('Serie Geometrica para $r=$%.2f y $N=$%.1f ' %(r,N))
```

```python
r=1; N=50
x3=geometrica(r,N)
plt.plot(x3,'o-',color='k')
plt.title('Serie Geometrica para $r=$%.2f y $N=$%.1f ' %(r,N))
```

```python
r=2; N=50
x4=geometrica(r,N)
plt.plot(x4,'o-',color='r')
plt.title('Serie Geometrica para $r=$%.2f y $N=$%.1f ' %(r,N))
```

## **Sobre la convergencia:**
<p align="justify">
De las primeras dos graficas, donde $r=1/2$ y $r=1/5$, que cumplen $\left|r\right|$<$1$ se puede observar que al tender $N$ a valores grandes, la sumatoria tiende a converger a un valor particular fijo, en este caso $2$  y $1.25$ respectivamente. Luego la serie geometrica para estos valores de $r$ es convergente.</br> </br>
Por su parte, para $r=1$ y $r=2$ ya no se cumple la condicion $\left|r\right| $<$1$ y podemos observar en la grafica que la suma crece monotonamente linealmente y potencialmente respectivamente, por lo que estas series nunca se estabilizaran un valor fijo, y por tanto divergerán.
