```python
# Valores a y b
ab = {'H2': (0.01945, 0.022),
      'H2O': (0.56539, 0.031),
      'N2': (0.13882, 0.039),
      'O2': (0.13983, 0.032),
      'CO2': (0.37186, 0.043)}

v = np.linspace(0.001, 0.5, 50)/n # volumen espefico
t = np.linspace(0, 300, 300)

def presion(a, b, v, T):
    return (R*T)/(v - b) - a/(v**2)


plt.figure(figsize=(6, 4))

for t_i in t:
    p = [presion(ab['H2'][0], ab['H2'][1], v_i, t_i) for v_i in v]
    plt.plot(v, p )

plt.xlabel('[m$^3$/mol]')
plt.ylabel('[Pa]')
plt.title(f'Diagrama Pv, ecuacion van der Waals')
plt.grid()
```

$v_C$ y $T_C$ son la presión crítica, el volumen molar crítico y la
temperatura crítica, entonces la presión reducida PR, el volumen molar reducido vR y la temperatura reducida TR están definidas como $ P_R = \frac{P}{P_C}$ , $v_R =\frac{v}{v_C}$ , y $T_R = \frac{T}{T_C} $

**(a)** Demuestre que en términos de cantidades reducidas, la ecuación de van der Waals se puede escribir

$$(P_R + \frac{3}{v_R^2}) (v_R - \frac{1}{3}) =\frac{8}{3}T_R$$
<br>

Cuando la ecuación de van der Waals aparece de esta forma, las constantes a y b, que dependen del material, no aparecen explícitamente. De esta manera, todos los gases que obedecen la ecuación de van der Waals se pueden considerar que están en el mismo estado cuando los valores de $P_R$, $v_R$ y $T_R$ son los mismos (es decir, cada gas esta medido en unidades de sus valores particulares de $P_C$, $v_C$ y $T_C$). Este es el principio de los estados correspondientes, el cual es un principio de similaridad universal establecido
por van der Waals. <br><br>

> Partiendo de la ecuacion de van der waals:

$$ (P + \frac{a}{v^2} ) (v - b) = R T $$

> para encontrar el valor critico de la temperatura, vamos a ver cuando la primer derivada se anula:
$$ \frac{dP}{dv} = -\frac{RT}{(v - b)^2} + \frac{2a}{v^3} = 0$$
$$ T_c = \frac{2a(v - b)^2}{Rv^3}$$

> similarmente, para el volumen critico se emplea la segunda derivada
$$ \frac{d^2P}{dv^2} = \frac{2RT}{(v - b)^3} - \frac{6a}{v^4} = 0$$
$$ \frac{2RT_c}{(v_c - b)^3} = \frac{6a}{v_c^4} $$
$$ T_c = \frac{6a(v_c - b)^3}{2Rv_c^4} =  \frac{2a(v_c - b)^2}{Rv_c^3}$$
$$  \frac{3(v_c - b)}{2} =  v_c$$
$$  v_c =  3b$$

> volviendo a la temperatura critica:
$$ T_c = \frac{2a(2b)^2}{R(3b)^3}$$
$$ T_c = \frac{8a}{R*27b}$$


> con la presion y temperatura criticos, encontramos la presion critica en la ec de van der waals:

$$ P_c  = \frac{R T_C}{(v_c - b)} - \frac{a}{v_c^2}$$
$$ P_c  = \frac{R 8a}{R*27b(v - b)} - \frac{a}{(3b)^2}$$
$$ P_c  = \frac{4a}{27b^2} - \frac{a}{9b^2}$$
$$ P_c  = \frac{a}{27b^2}$$

> Ahora con las tres variables criticas, se manipula para llegar a la expresion reducida:
$$ \frac{R T_c}{P_c v_c}= \frac{27 b^2 * R*8*a }{a *3b*27R*b} = \frac{8}{3} $$

> en van der waals:
$$ \frac{P_c}{v_C}(P_R v^2+ \frac{a}{v_R^2 P_c} ) (v_R - \frac{b}{v_c}) = R T$$
$$ (P_R v_c^2+ \frac{a}{v_R^2 P_c} ) (v_R - \frac{1}{3}) = \frac{v_C}{P_c}R T_c T_R$$
$$ (P_R + \frac{a}{(3b)^2 v_R^2 (a/27b^2)} ) (v_R - \frac{1}{3}) = \frac{8}{3} T_c$$
$$ (P_R + \frac{3}{v_R^2} ) (v_R - \frac{1}{3}) = \frac{8}{3} T_c$$


**(b)** Utilizando Python, realice un diagrama de tres curvas para $P_R$ como función de $v_R$, uno para $T = ½T_C$, otro para $T = T_C$, y otro para $T= 2T_C$. ¿Qué ocurre físicamente cuando la ecuación indica tres posibles valores de vR para un solo valor de $P_R$ y $T$?

```python
T_r = [0.5, 1, 2] # temperaturas reducidas dadas

v_r = np.linspace(0.1, 0.6, 60) # volumen reducido

def P_r(v_r, T_r): # P_r presion reducida
    return (8*T_r)/(3*v_r - 1) - 3/v_r**2


plt.figure(figsize=(6, 4))
for T_r_i in T_r:
    plt.plot(v_r, P_r(v_r, T_r_i), label=f'T$_r$ = {T_r_i}')

plt.xlabel('V$_r$')
plt.ylabel('P$_r$')
plt.title('Diagrama P$_r$ V$_r$ van der Waals (Cantidades Reducidas)')
plt.legend()
plt.grid()
```

#### **Posible Analisis Fisico**

Los puntos donde se da este comportamiento, es decir tienen el mismo valor de
$P_r$ y $T_r$ para un volumen reducido especifico son llamados "estados correspondientes" y en general poseen comportamientos similares, siempre que se escriban en términos de sus cantidades reducidas. En este caso,  se da para aproximadamente la presion reducida $P_r=0$, lo que implica que hay tres posibles estados para este gas que tienen la misma presión y temperatura reducidas.
