### **Eficiencia en maquinas Termicas**

$$ \eta =  \frac{W}{Q_H} = 1- \frac{Q_L}{Q_H} $$


Teniendo en cuenta que para una maquina de carnot ouna reversible se cumple que la eficiencia siempre es masyor o igual que la de una irreversible $ η_R ≥ η_I$. Para poner la eficiencia en terminos de la temperatura en cada uno de los siguientes ciclos se hace uso de la primer ley de la termodinamica $ΔU=Q+W$ y se explota el hecho particular de cada proceso, si es isotermico, isobarico, o adiabatico donde funcionan las relaciones vistas en clase, de este modo se pueden obtener las siguientes ecuaciones para la eficiencia:

#### (a) Ciclo de **Otto** <br>
<img src=https://i.postimg.cc/bY5Z8k6k/otto.png width="329">

* 2 ciclos adiabaticos (1-2 compres) (3-4 expansi)
* 2 ciclos isocoricos (2-3) (4-1)

$$ \eta  = 1 - \frac{T_4-T_1}{T_3-T_2} = 1- \frac{T_1}{T_2} = 1- \frac{T_4}{T_3} $$


<br>

#### (b) Ciclo de **Diesel** <br>
<img src=https://i.postimg.cc/KcT1G4CW/diesel.png width="300">

* 2 ciclos adiabaticos (1-2 compres) (3-4 expansi)
* 1 ciclo isobarico (2-3 expansi)
* 1 ciclo isocorico (4-1)

$$ \eta  = 1 - \frac{1}{γ} \frac{T_4-T_1}{T_3-T_2} $$

es semejante a la eficincia en otto a exepcion del factor $γ$ que mientras que sea mas grande, la eficiencia será mayor que la de Otto
<br>
#### (c) Ciclo de **Joule** o de **Brayton** <br>
<img src=https://i.postimg.cc/6qbjbYwH/joule.png width="329">

* 2 ciclos isobaricos  (2-3 expansi) (4-1 compres)
* 2 ciclos adiabaticos  (1-2 compres) (3-4 expansi)

$$ η = 1 - \frac{T_1}{T_2}$$

<br>

#### (d) Ciclo de **Stirling** <br>
<img src=https://i.postimg.cc/Y0NghSDQ/stirling.png width="329">

* 2 ciclos isocoricos (2-3 ) (4-1)
* 2 ciclos  (1-2 ) (3-4 )


$$ η = 1 - \frac{T_3}{T_2}$$


```python
# librerias
import numpy as np
import matplotlib.pyplot as plt
```

suponiendo las siguientes temperaturas aleatorias en el rango $[1,500]k$, aplicando los condicionales apropiados para que no se viole la **2 ley de la termodinamica** que la eficiencia sea del $100%$ o mayor, y que para efectos practicos no sea negativa, se tiene el siguiente codigo. (se ejecuta hasta que las condiciones se cumplan, y los valores se impriman, si no se imprimen los valores se debe a que la seleccion aleatoria no cumple las condiciones mencionadas anteriormente)

```python
# parametro gamma c_p/c_v
γ_mono  = 5/3
γ_diat  = 7/5
```

```python
# temperaturas
T_1 = np.random.uniform(1,500)
T_2 = np.random.uniform(1,500)
T_3 = np.random.uniform(1,500)
T_4 = np.random.uniform(1,500)


if (T_4-T_1)/(T_3-T_2)<1:
  # otto
  η_otto = 1 - (T_4-T_1)/(T_3-T_2)  # 1- T_1/T_2 #= 1 - T_4/T_3

  # diesel
  η_dies = 1- (γ_mono)**-1 * (T_4-T_1)/(T_3-T_2)

# joule
if T_1/T_2<1:
  η_joule = 1- T_1/T_2

# stirling
if T_3/T_2<1:
  η_stir = 1- T_3/T_2 #= T_4/T_1

# condicional de que la eficiencia siempre es menor que 1 y mayor que 0
if 1>η_otto>0 and 1>η_dies>0 and 1>η_joule>0 and 1>η_stir>0:
  print('Eficiencias: \n')
  print(' Ciclo Otto: ', η_otto)
  print(' Ciclo Diessel: ', η_dies)
  print(' Ciclo Joule: ', η_joule)
  print(' Ciclo Stirling: ', η_stir)
```

#### **Conclusion y Discusiones**

lo primero a observar, y que resulta mevidente de las ecuaciones de la eficiencia para el ciclo de Otto (motores convensionales) siempre es menor que la ecficiencia del ciclo Diessel (que requiere conmbustible especial a diferencia de los motores comunes), lo cual se debe al cambio del proceso (1-2) de isocorico en Otto a isobarico en Diessel, este cambio hace que surja el parametro $γ$ en el denominador del segundo termino (igual para ambos casos), entonces este parametro, entre sea mas grande, como pasando de gas monoatomico a diatomico, hará que la eficiencia en Diessel sea aun mas superior que en Otto. <br>

Cabe resaltar que este parametro $γ$ solo aparece en el ciclo Diessel debido a que combina proceso isobarico e isocorico, de donde surgen las capacidades calorificas a presion y volumen constante con las que se define tal parametro. por su parte el Diessel es el unico de estos 4 ciclos que combina estos dos procesos, por eso el cambio de la composicion del gas solo afectará a este proceso, y dejará invariante a los otros 3. <br>

A su vez, para los otros dos ciclos faltantes, el de Joule y Stirling no se podria afirmar cual de los dos sea mas eficiente que el otro, pues la dependencia de los valores particulares de la temperatura hace que no tengan una relacion clara, a menos que se lograra establecer una relacion entre la temperatura 3 y la 1. <br>
