```python
# librerias
import numpy as np
import matplotlib.pyplot as plt
from astropy import constants
```

**6.35.** La ecuación de estado de van der Waals está dada por:
$$ (P + \frac{a}{V^2} ) (V - b) = R T $$
$$ P = \frac{R T}{V - b} - \frac{a}{V^2}$$


Asuma que un gas de van der Waals experimenta una expansión **isotérmica**, a la temperatura T, desde un volumen V1 hasta V2. Calcule el cambio en las siguientes energias. Las expresiones deben estar dadas en términos de $T, V_1, V_2$ y de las constantes $R$, $a$ y $b$. <br><br>







* **(b)** La energía interna ($U$) <br>
>  $$ dU = T dS - P dV $$
>  $$ ΔU = \int_{S_0}^{S_f} T dS -   \int_{V_0}^{V_f} P dV  $$
>
>  usando 1r ec TdS: $ T dS = C_v dT+ T (\frac{∂P}{∂T})_{V} dV = T (\frac{∂P}{∂T})_{V} dV = \frac{R T }{V-b} dV$ <br>
>  
>  $$ ΔU = \int_{V_0}^{V_f} \frac{R T }{V-b} dV -   \int_{V_0}^{V_f} (\frac{RT}{V-b}- \frac{a}{V^{2}} ) dV $$
>
>  $$ ΔU = RT \ln \frac{V_f-b}{V_0-b} -  RT \ln \frac{V_f-b}{V_0-b} - \frac{a}{V_f} + \frac{a}{V_0}  $$
>
>  $$ ΔU = - \frac{a}{V_f} + \frac{a}{V_0}  $$



* **(a)** La energía libre de Helmholtz ($A$) <br>
> $$ dA = -S dT  -P dV $$
> $$ dA =  -P dV $$
> $$ ΔA = -\int_{V_0}^{V_f} (\frac{RT}{V-b}- \frac{a}{V^{2}} ) dV  $$
> $$ ΔA =  -RT \ln \frac{V_f-b}{V_0-b} - \frac{a}{V_f} + \frac{a}{V_0} $$



* **(c)** La energía libre de Gibbs ($G$)<br>
>  $$ dG = -S dT + V dP $$
>  $$ dG = V dP $$
>
> Usando: $ dP = ( -\frac{RT}{(V-b)^2} +  \frac{2a}{V^3} )dV$ de van del Waals
>
 > $$ ΔG = \int_{V_0}^{V_f}( -\frac{RT V}{(V-b)^2} +  \frac{2a}{V^3} )dV $$
 > $$ ΔG = -RT \int_{V_0}^{V_f} \frac{u+b}{u^2} du +  \int_{V_0}^{V_f} \frac{2a}{V^3} dV $$

 > $$ ΔG = -RT \ln \frac{V_f-b}{V_0-b} + \frac{bRT}{V_f-b} - \frac{bRT}{V_0-b} -  \frac{a}{V_f^2} + \frac{a}{V_0^2} $$
>  
  > $$ ΔG = -RT \ln \frac{V_f-b}{V_0-b} + bRT(\frac{1}{V_f-b} - \frac{1}{V_0-b}) + a(-\frac{1}{V_f^2} + \frac{1}{V_0^2}) $$

<br>

* **(d)** La entalpía. ($H$) <br>
>  $$ dH = T dS + V dP $$
>
> usando los resultados anteriores, para el caso isotermico se puede decir:
>
 > $$ ΔH = RT \ln \frac{V_f-b}{V_0-b} - RT \ln \frac{V_f-b}{V_0-b} + bRT(\frac{1}{V_f-b} - \frac{1}{V_0-b}) + a(-\frac{1}{V_f^2} + \frac{1}{V_0^2})  $$
>
 > $$ ΔH = bRT(\frac{1}{V_f-b} - \frac{1}{V_0-b}) + a(-\frac{1}{V_f^2} + \frac{1}{V_0^2})  $$
>

#### **Diagramas**

Para distintos valores de $T$ y de $V_1$, realice **diagramas** de $ΔA$, $ΔU$, $ΔG$ y $ΔH$ versus $V_2$ para distintos gases,

por lo menos tres (consulte valores reportados de las constantes $a$ y $b$). Interprete los resultados obtenidos.

```python
# constantes
R = constants.R.value # R [J/K.mol]

# Valores de a y b para: gas ideal, N2, O2 y CO2
names_gas = ['ideal', 'N2', 'O2', 'CO2']
a = [0, 1.39, 1.36, 3.59]        # [L^2 atm/mol^2]
b = [0, 0.0391, 0.0318, 0.0427]  # [L/mol]

# parametros
T = 300  # [K]  Temperatura constante
V0 = 0.7  # [L]  Volumen inicial
V2 = np.linspace(0.9, 1.5, 100)  # [L] rango de volumen
```

```python
# cambios de energias de cada gas
delta_A = [] # Helmholtz A
delta_U = [] # interna U
delta_G = [] # Gibbs G
delta_H = [] # Entalpia H

for i in range(len(a)):  # para cada gas
    ΔA = -R*T*np.log((V2 - b[i])/(V0 - b[i])) + a[i]*(-1/V2 + 1/V0)
    ΔU = a[i]*(-1/V2 + 1/V0)
    ΔG = -R*T*np.log((V2 - b[i])/(V0 - b[i])) + b[i]*R*T*(1/(V2-b[i]) - 1/(V0-b[i])) + a[i]*(-1/(V2**2) + 1/(V0**2))
    ΔH = b[i]*R*T*(1/(V2 - b[i]) - 1/(V0 - b[i])) + a[i]*(-1/(V2**2) + 1/(V0**2))

    delta_A.append(ΔA)
    delta_U.append(ΔU)
    delta_G.append(ΔG)
    delta_H.append(ΔH)

delta = [delta_A,delta_U,delta_G,delta_H]
Δ_names = ['ΔA','ΔU','ΔG','ΔH']

# Graficar ΔA, ΔU, ΔG, ΔH versus V2 para los tres gases
fig, ax = plt.subplots(figsize =(14,4), ncols=4, nrows= 1)

for i in range(len(a)):
  ax[i].set_title(f'{Δ_names[i]} vs $V_2$ a {T}K')
  ax[i].plot(V2, delta[i][0], label='ideal')
  ax[i].plot(V2, delta[i][1], label='N2')
  ax[i].plot(V2, delta[i][2], label='O2')
  ax[i].plot(V2, delta[i][3], label='CO2')

  ax[i].set_xlabel('[L]')
  ax[i].legend()
  ax[i].grid()

ax[0].set_ylabel('[J]')
plt.show()
```

### **Conclusion**

La energia de Helmholtz A, en particular su cambio (ΔA) disminuye a medida que el volumen $V$ aumenta para cualquier de los gases tomados en cuenta, incluso el ideal. Luego a mayor volumen la energia libre de Helmholtz disminuye, *lo que es consistente con la tendencia de que los gases se expandan espontáneamente a volúmenes mayores.*


El cambio en la energía interna ΔU aumenta a medida que el volumen aumenta, es decir mientras los gases se expanden. mientras que para el gas ideal el cambio es nulo.


El cambio en la energia libre de Gibbs al igual que el cambio en la entalpia tambien tiende a disminuir, *lo que da indicios de como las fuerzas repulsivas y atractivas entre las particulas interactuan durante la expansión isotermica.*
