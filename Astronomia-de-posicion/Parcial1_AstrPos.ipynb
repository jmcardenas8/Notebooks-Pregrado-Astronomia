```python
# ALGUNOS PARAMETROS DE LA TIERRA

e = 0.081819    #exentricidad
f = 1/298.257   #achatamiento
a = 6378140     #semieje-mayor [m]
b = 6356755     #semieje-menor [m]
radiot = 6371   #radio terreste[km]

rad= np.pi/180  #grados a rads
```

```python
# TEOREMAS DE LA TRIGONOMETRIA ESFERICA

def tma_sin(a,c,C):       #teorema de seno, para angulo (A)
  return np.arcsin( (np.sin(C)*np.sin(a))/np.sin(c) )

def cos_angulos(A,B,c):   #teorema de coseno para angulos (C)
  return np.arccos( -np.cos(A)*np.cos(B) + np.sin(A)*np.sin(B)*np.cos(c) )

def cos_lados(a,b,C):     #teorema de coseno para lados (c)
  return np.arccos( np.cos(a)*np.cos(b) + np.sin(a)*np.sin(b)*np.cos(C) )

def cos_lados_mod(a,b,c): #tma coseno modificado para obtener 1 angulo a partir de 3 lados
  return np.arccos( (np.cos(c)-np.cos(a)*np.cos(b)) / (np.sin(a)*np.sin(b))  )  #angulo (C)

def hav_sin(a,b,C):       #teorema de haverseno (variante del tma cos lados con +precision: angulos pequeños) (c)
  return   2* np.arcsin( ( np.sin((a-b)/2)**2  +  np.sin(a)*np.sin(b)*np.sin(C/2)**2   )**0.5 )
```

# **B ) Transformacion de Coordenadas:** Geodesicas a Geocentricas


<p align="justify">

Funcion que dado el angulo de `latitud geodesica` en radianes y la altura $h$ devuelve una lista donde el primer indice  corresponde a la `latitud geocentrica` en radianes, el segundo a la `distancia geocentrica` mas la altura $h$ en metros y el tercero a la misma distancia pero sin sumar $h$. Para el calculo de cada uno se emplearon las formulas de transformacion de coordenadas geodesicas-geocentricas.
</p>

```python
def geodesic_2_centric(θ, h):        # Funcion de geodesicas a geocentricas
  θ_prim = 0                         # intento de limpiar los valores
  θ_prim = np.arctan( (1-e**2) * np.tan(θ) ) # latitud geocentrica
  p = a *( (1-e**2*np.sin(θ)**2 *(2-e**2)) / (1-e**2*np.sin(θ)**2) )**0.5
  p_prim = p+h                       # distancia geocentrica con altura
  return θ_prim, p_prim, p           # latitud en radianes
```

```python
#Evaluando los 3 puntos en la funcion (con h=0)
Φ_pA, p_A, pa = geodesic_2_centric ( ΦA , 0 )
Φ_pB, p_B, Pb = geodesic_2_centric ( ΦB , 0 )
Φ_pC, p_C, Pc = geodesic_2_centric ( ΦC , 0 )

#latitud geocentrica de los puntos
print('Latitudes Geocentricas de los 3 puntos:')
print(f" Φ'A= {Φ_pA/rad:.3f}°")
print(f" Φ'B= {Φ_pB/rad:.3f}°")
print(f" Φ'C= {Φ_pC/rad:.3f}°")
```

# **D ) Distancia entre 2 puntos cualesquiera sobre la Tierra**

Funcion en la cual entran las `latitudes y longitudes` de los dos puntos en `radianes`, la funcion toma como vertices del triangulo esferico a estos dos puntos junto al polo norte terrestre. Empleando el teorema de coseno para lados o su version mas precisa del Semiverseno para angulos pequeños (en este caso se usó el semiverseno pues algunos algunos angulos son relativamente pequeños) se encuentra el lado del triangulo que separa los puntos deseados. Con el cual finalmente la funcion retorna una lista donde el primer indice es la `distancia` entre los dos puntos en km, y el segundo indice es el lado del triangulo esferico que separa los dos puntos en radianes.

<img src=https://i.postimg.cc/SRbS5qvF/t9.jpg width="450">



```python
def distancia( λ1, Φ1, λ2, Φ2, h):

  #Φ_p1 = geodesic_2_centric( Φ1 , 0 )[0] #transforma geodesicas a geocentricas
  #Φ_p2 = geodesic_2_centric( Φ2 , 0 )[0] # no necesario

  C = np.abs( λ1 - λ2 )            #val.abs garantiza para cualquier longitud(+,-)
  a = np.abs( np.pi/2 - Φ2 )       #se toma el triangulo A,B, PNT
  b = np.abs( np.pi/2 - Φ1 )
  print('parametros en radianes:')
  print(f' C= {C:.3f}')
  print(f' a= {a:.3f}')
  print(f' b= {b:.3f}')

  if C < np.pi:                    #condicion angulos y lados deben ser <180
    c = hav_sin (a,b,C)            #tma coseno lados, version semiversin
    dist = (radiot+h)*c
    print(f' c= {c:.4f}')
    print('La distancia entre los dos puntos es: %.2f km' %dist )

    #adicionalmente para resolver todo el triangulo solo faltan A y B:
    #A = cos_lados_mod(c,b,a); B = cos_lados_mod(a,c,b)
    #print(f' A= {A:.3f}') ; print(f' B= {B:.3f}')

    return dist, c

  else:
    C_new = 2*np.pi - C            #conjugado de c
    c = hav_sin (a,b,C_new)        #tma coseno lados, version semiversin
    dist = (radiot+h)*c
    print('Como C>180, se toma su conjugado: C= %.2f' %C_new )
    print(f' c= {c:.4f}')
    print('La distancia en km entre los dos puntos es: %.2f km' %(radiot * c) )

    #adicionalmente para resolver todo el triangulo solo faltan A y B:
    #A = cos_lados_mod(c,b,a); B = cos_lados_mod(a,c,b)
    #print(f' A= {A:.3f}') ; print(f' B= {B:.3f}')

    return dist, c                #distancia [km] y c[rads]
```

# **E ) Area de Suramerica**
</br>

Para estimar el area de Suramerica se hace uso de la formula para el area superficial de triangulos esfericos, dada por la ecuacion: $^{[1,2]}$

$$ S= \epsilon ^{'}* R_{\oplus}^{2} $$

Donde $S$ es el area del triangulo, $R_{\oplus}$ es el radio promedio de la tierra, y $\epsilon ^{'}$ es el exceso esferico en radianes dado por: $$\epsilon ^{'}=(A+B+C-\pi)$$  
Con $A,B,C$ los angulos del triangulo esferico en radianes.

<br>

<p align="justify">
En este caso tenemos los 3 lados del triangulo esferico ($c1,c2$,y $c3$ encontrados en el punto anterior), luego el triangulo se puede resolver para encontrar sus 3 respectivos angulos internos($C1,C2$ y $C3$), para lo cual se hizo uso de una modificacion del teorema del coseno para lados, despejando el unico factor que contiene un angulo en terminos de los tres lados.</p>
<br>

<img src=https://i.postimg.cc/FRrmfTGw/t8.jpg width="600">


**Bibliografia:**
1.  Daniel J. Briones (2017). Trigonometria en la Esfera. Pg-6.
2.  Universidad Politecnica de Madrid. (2008). Trigonometria Esferica Basica. Pg-14.

<p style='text-align: justify;'> </p>



```python
def angulos_lados(a,b,c):      #funcion dados 3 lados retorna los 3 angulos(usa la modificacion del tma coseno)
  A= cos_lados_mod(b,c,a)      #para el angulo opuesto a 'a', es decir 'A'
  B= cos_lados_mod(a,c,b)
  C= cos_lados_mod(a,b,c)
  return A,B,C                 #angulo correspondiente a (a,b,c) respectivamente
```

```python
def area_trig_esferic (A,B,C): # Fucion area de un triangulo esferico
  eps = A+B+C -np.pi           # exceso esferico
  s = eps * radiot**2          # Area
  return s, eps
```

```python
# Evaluando la funcion con los 3 lados para obtener los 3 angulos
C1,C2,C3 = angulos_lados(c1,c2,c3)

# Evaluando los angulos en la funcion del area
s , eps = area_trig_esferic(C1,C2,C3)

print('-----------------------------------------------------------')
print(f'El exceso esferico del triangulo es {e:.5f}')
print('-----------------------------------------------------------')
print(f'El area del triangulo de Suramerica es = {s:.2f} km²')
print('-----------------------------------------------------------')
print(f'El perimetro del triangulo es = {radiot*(c1+c2+c3):.2f} km')
```

# **F ) Comparacion con el Area real de Suramerica y error**
</br>

De la literatura se tiene que el territorio de Suramerica tiene una superficie de al rededor de los 17 millones ochocientosmil kilometros cuadrados. Dependiendo de la fuente este valor varia levemente; de *wikipedia* se tiene $17'855.914,00$ $km^{2}$  $^{[1]}$, de *EcuRed* una superficie de $17'819.100,0$ $km^{2}$  $^{[2]}$, de *New world encyclopedia* de $17'840.000,0$ $km^{2}$  $^{[3]}$. # **G ) Posibles mejoras en el calculo**

<p align="justify">
Como se muestra en las figuras del principio, con un solo triangulo no se puede inscribir la figura del continente completamente, y si se hace, se agregarán muchas partes oceanicas que causarán que el error con respecto a la medida real sea aun mayor. Una mejor aproximacion podria ser la creacion de no solo un triangulo esferico sino varios, los suficientes como para poder modelar y acomodarse a las costas reales del continente.
Para esto se necesitaria una tabla de datos de las coordenadas de los puntos, y con ciclos "for" repetir el proceso hecho anteriormente para cada uno de los triangulos, al final tomando el area total como la suma de la superficie de cada triangulo. </p>
